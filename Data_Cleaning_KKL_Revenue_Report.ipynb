# Objective

The objective of this task is to inspect the Revenue Report (2019–2021) for formatting inconsistencies, clean the dataset, convert month names into numeric values, identify data anomalies, and prepare a cleaned dataset suitable for visualization and further analysis.

```python
import pandas as pd
import numpy as np
```

# Load Dataset

The Revenue Report dataset is loaded into the Python environment using the Pandas library.

```python
df = pd.read_excel("S_Data_KKL Revenue Report_19-21_1.xlsx")

df.head()
```

# Dataset Overview

The structure of the dataset was examined to understand its dimensions, column names, data types, and completeness before performing any cleaning operations.

```python
print(df.shape)

df.info()
```

# Activity 1 – Inspection for Formatting Errors

The dataset was inspected to identify:
- Missing values
- Duplicate records
- Blank records
- Inconsistent formatting
- Incorrect data types

```python
df.isnull().sum()
```

## Key Findings

- Dataset contains **54 duplicate records**.
- Several columns contain missing values.
- The **Accounts** column has approximately **98.9% missing values**.
- Month names are consistently formatted.

# Data Cleaning

## Step 1: Remove Duplicate Records

Duplicate records can affect the accuracy of data analysis and visualization. Therefore, duplicate rows were identified and removed from the dataset.

```python
print("Rows before removing duplicates:", df.shape)

duplicates = df.duplicated().sum()
print("Duplicate rows found:", duplicates)

df = df.drop_duplicates()

print("Rows after removing duplicates:", df.shape)
```

## Step 2: Remove Incomplete Records

Rows containing missing values in key columns such as Fiscal Month, Shipped Date, Customer ID, and Sales Amount were removed to improve data quality.

```python
df = df.dropna(
    subset=[
        'Fiscal Month',
        'Shipped Date',
        'Customer ID',
        'Sales Amount'
    ]
)

print("Dataset Shape after removing incomplete rows:")
print(df.shape)

print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

## Step 3: Standardize Text Values

Text columns were standardized by removing leading and trailing spaces and converting selected fields to Title Case to ensure consistency.

```python
# Remove leading/trailing spaces
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = df[col].str.strip()

# Standardize capitalization
columns_to_standardize = [
    'Fiscal Month',
    'Sales Channel',
    'Ship To Country',
    'Product Type',
    'Company'
]

for col in columns_to_standardize:
    df[col] = df[col].str.title()

print("Text formatting standardized successfully.")
```

## Data Cleaning Summary

The following data cleaning operations were completed:

- Removed 54 duplicate records.
- Removed incomplete records with missing values in key fields.
- Standardized text formatting by trimming extra spaces and applying consistent capitalization.
- Identified the **Accounts** column as a major anomaly with 98.9% missing values.
- Prepared the dataset for month conversion and visualization.

# Activity 3 – Convert Month Names into Numeric Values

To facilitate data analysis and visualization, the Fiscal Month column was converted from month names to their corresponding numeric values (January = 1, February = 2, …, December = 12).

```python
# Dictionary to map month names to numbers
month_map = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}

# Create a new column with month numbers
df['Month_Number'] = df['Fiscal Month'].map(month_map)

print("Month names successfully converted into numeric values.")
```

```python
# Display converted values
df[['Fiscal Month', 'Month_Number']].head(15)
```

# Activity 4 – Document Anomalies

The dataset was analyzed to identify anomalies that may affect data quality and reliability.

The anomalies observed during inspection and cleaning are summarized below.

```python
# Summary of anomalies

anomalies = {
    "Duplicate Records": 54,
    "Incomplete Records Removed": 2,
    "Missing Ship To Country": df['Ship To Country'].isnull().sum(),
    "Missing Accounts": df['Accounts'].isnull().sum()
}

anomaly_df = pd.DataFrame(
    list(anomalies.items()),
    columns=["Anomaly", "Count"]
)

anomaly_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(anomaly_df["Anomaly"], anomaly_df["Count"])

plt.title("Summary of Data Anomalies")
plt.xlabel("Anomaly")
plt.ylabel("Count")

plt.xticks(rotation=20)

plt.show()
```

## Key Findings

- 54 duplicate records were identified and removed.
- Two incomplete records were removed from the dataset.
- The **Accounts** column contains a high percentage of missing values (2556 records).
- The **Ship To Country** column still contains 33 missing values.
- Month names were successfully converted into numeric values for easier analysis.

# Activity 5 – Prepare Cleaned Dataset

The cleaned dataset was exported in Excel format. This dataset is now ready for visualization and can be used by the next team member for revenue reconciliation and analysis.
